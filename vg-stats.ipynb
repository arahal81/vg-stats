```python
import pandas as pd
df = pd.read_csv('./vgsales.csv')
df
```

1- Which company is the most common video game publisher?


```python

most_common_publisher =df['Publisher'].value_counts().idxmax()
most_common_publisher
```

2- Whatâ€™s the most common platform?


```python
most_common_platform =df['Platform'].value_counts().idxmax()
most_common_platform
```

3- What about the most common genre?


```python
most_common_genre =df['Genre'].value_counts().idxmax()
most_common_genre
```

4- What are the top 20 highest grossing games?

```python

top_twenty_highest_grossing_games = df[['Name','Global_Sales']].sort_values('Global_Sales', ascending=False).head(20)
top_twenty_highest_grossing_games
```

5- For North American video game sales, whatâ€™s the median?

```python
na_median_sales = df['NA_Sales'].median()
na_median_sales
```

5.1- Provide a secondary output showing ten games surrounding the median sales output

```python
ten_median_na_seller_names = df[df['NA_Sales'] == na_median_sales][['Name']].head(10)
ten_median_na_seller_names
```

5.2- assume that games with same median value are sorted in descending order

```python
median_ten_na_seller_names_descending = (df[df['NA_Sales'] == na_median_sales][['Rank','Name']].sort_values('Rank',ascending=False))[['Name']].head(10)
median_ten_na_seller_names_descending
```

6- For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?

```python

result= (df['NA_Sales'].head(1)-df['NA_Sales'].mean())/df['NA_Sales'].std()
result
```

7- The Nintendo Wii seems to have outdone itself with games. How does its average number of sales compare with all of the other platforms?

```python

sales_compare  = (df['Global_Sales'].mean(),df[df['Platform'] == 'Wii']['Global_Sales'].mean())
sales_compare 
```
